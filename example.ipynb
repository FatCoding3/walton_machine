Place the `walton.py` file in the same folder with this notebook or you have to config your import command.

```python
import matplotlib.pyplot as plt
from walton import WaltonMachine
```

# Playground

First, you have to create a `WaltonMachine` instance.

```python
# Create an instance of walton machine
walton_machine = WaltonMachine(
  # Number of capacitors in the upper.
  # Must be integer
  upper_capacitors=10, 

  # The voltage of the power source.
  voltage=200
)
```

After a machine is created, you can view the system via `WaltonMachine.print_system()`.

```python
walton_machine.print_system(
  # Step of the machine wanted to view.
  step=0
)
```

You update the system with `WaltonMachine.update()`.

```python
# Apply next switch 
walton_machine.update(
  # You can update many times at once.
  update_times=1001,

  # You can choose to view the final state or not.
  # Default is True
  print_system=False
)
```

```python
# Print the final state of the machine
walton_machine.print_system(step=-1)
```

You can plot the summary voltage, max possible voltage or voltage of any capacitor in the function of step via `WaltonMachine.plot_history()`.

```python
# You can modify any properties of plt before and after plotting
plt.title('System voltage')
plt.xlim(0, 100)

# Plot voltage
walton_machine.plot_history(
  # Step range you want to plot
  # If this isn't provided, the system will print all steps
  step_range=[0, 1000],

  # The system prints only odd step if this is True
  # Default is True
  only_odd_step=False,

  # Plot the possible max voltage or not
  # Default is True
  have_max_voltage=True,

  # Plot the summary voltage or not
  # Default is True
  have_sum_voltage=True,

  # Choose upper capacitors to plot their voltage
  # Default is []
  upper_capacitors=[2, 5],

  # Choose lower capacitors to plot their voltage
  # Default is []
  lower_capacitors=[2],

  # If this is True, the voltage of all capacitors will be plot
  # Default is False
  all_capacitors=False
)

plt.show()
```
